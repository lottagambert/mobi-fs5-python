**[Task 2.1.]()** Implement Otsu thresholding as a re-usable function:

```python
imgf = plt.imread("data/NIH3T3/im/dna-0.png")
img8 = (imgf * 255).round().astype(np.uint8)

def otsu_thresholding(imgf):
    img8 = (imgf * 255).round().astype(np.uint8)
    mög_T = np.inf
    T_fin = 0

    for T in range (1, 256):
        class_1 = img8[img8 < T ] #[0, T-1]
        class_2 = img8[img8 >= T] #[T, 255]

        N_1 = len(class_1)
        N_2 = len(class_2)

        sigma_1 = np.var(class_1)
        sigma_2 = np.var(class_2)

        min_var_T = N_1 * sigma_1 + N_2 * sigma_2


        if min_var_T < mög_T:

            mög_T = min_var_T
            T_fin = T
            
    seg = (img8 >= T_fin).astype(bool)
       
    return seg



```

**[Task 2.2.]()** Implement a re-usable function to compute the Dice coefficient:

```python
def compute_dice(bin1, bin2):
    assert bin1.dtype == bool
    assert bin2.dtype == bool
    
    intersection = np.logical_and(bin1, bin2) 
    
    dice = 2 * intersection.sum() / (bin1.sum() + bin2.sum())

    return dice
```

```python
imgf = plt.imread("data/NIH3T3/im/dna-0.png").astype(bool)
T = otsu_thresholding(imgf)
print(T)

T_bin = T.astype(bool)
Dice = compute_dice(T_bin, imgf)
print(Dice)
```

```python
#Write a for-loop which iterates the sequence i = 28, 29, 33, 44, 46, 49 and
#(a) loads the i-th image via plt.imread(f’data/NIH3T3/im/dna-{i}.png’),
#(b) loads the corresponding ground truth from f’data/NIH3T3/gt/{i}.png’,
#(c) performs Otsu thresholding,
#(d) computes the Dice coefficient of the segmentation result.


dice_list = []

for i in (28, 29, 33, 44, 46, 49):
    i_th = plt.imread(f'data/NIH3T3/im/dna-{i}.png')
    gt_i_th = plt.imread(f'data/NIH3T3/gt/{i}.png').astype(bool)

    seg1 = otsu_thresholding(i_th)
    dice1 = compute_dice(gt_i_th, seg1)
    print(dice1)
    dice_list.append(dice1)

print(np.mean(dice_list))
```
