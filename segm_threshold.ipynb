```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
img = plt.imread("data/NIH3T3/im/dna-0.png")
img.shape
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.figure(figsize=(12, 6))
plt.imshow(img)
plt.colorbar()
plt.title("DNA-0 Image")
plt.show()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
T = 0.10             
binary = (img >= T).astype(int)

plt.figure(figsize=(12, 6))
plt.imshow(binary, cmap='gray')
plt.title("Binary Image using Threshold T = 0.10")
plt.colorbar()
plt.show()
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
gt = plt.imread('data/NIH3T3/gt/0.png')

plt.figure(figsize=(12,6))
plt.imshow(gt, cmap='gray')
plt.title("Ground Truth")
plt.colorbar()
plt.show()
```

Compute and print the Dice coefficient:

```python
G = binary.astype(bool)
H = gt.astype(bool)

intersection = (G & H).sum()

dice = 2 * intersection / (G.sum() + H.sum())

print("Dice =", dice)

```
