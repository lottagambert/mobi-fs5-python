**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
from scipy.signal import convolve2d

# Bild laden und ggf. Graustufen
img = plt.imread('data/lena.png')
if img.ndim == 3:  # RGB  Graustufen
    img = img.mean(axis=2)

# Sobel-Kernels
Kx = np.array([[-1, 0, 1],
               [-2, 0, 2],
               [-1, 0, 1]])

Ky = np.array([[-1, -2, -1],
               [ 0,  0,  0],
               [ 1,  2,  1]])

# Horizontale und vertikale Ableitung
gx = convolve2d(img, Kx, mode='same', boundary='symm')
gy = convolve2d(img, Ky, mode='same', boundary='symm')
```

**[Task 2.3.]()** Test your implementations from above:

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(gx, cmap='gray')
plt.colorbar()
plt.title("Sobel Horizontal Gradient (gx)")

plt.subplot(1,2,2)
plt.imshow(gy, cmap='gray')
plt.colorbar()
plt.title("Sobel Vertical Gradient (gy)")

plt.show()
```

**[Task 2.4.]()** Compute the image gradient magnitude:

```python
# Magnitude des Gradienten
grad_mag = np.sqrt(gx**2 + gy**2)

# Ergebnis anzeigen
plt.figure(figsize=(6,6))
plt.imshow(grad_mag, cmap='gray')
plt.colorbar()
plt.title("Gradient Magnitude")
plt.show()


```
